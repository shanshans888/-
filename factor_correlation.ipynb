```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取CSV文件，假设文件名为data.csv，你可以根据实际情况修改
df = pd.read_csv('D:\factor_data.csv')
```

```python

# 将数据中的null值填补为0
df = df.fillna(0)

# 提取因子列
factor_cols = [col for col in df.columns if 'factor' in col]
factor_df = df[factor_cols]

# 计算因子之间的相关性
correlation_matrix = factor_df.corr()

# 绘制热力图
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Factors')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 筛选指定因子列
selected_factors = ['factor_1', 'factor_2', 'factor_4', 'factor_5', 'factor_8']
factor_df = df[selected_factors]

# 计算相关性矩阵
correlation_matrix = factor_df.corr()

# 绘制热力图（调整大小以适应5x5矩阵）
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1, 
            square=True, linewidths=.5, annot_kws={'size': 10})
plt.title('Correlation Heatmap (Factors 1,2,3,6,8)')
plt.tight_layout()  # 优化布局
plt.show()
```
